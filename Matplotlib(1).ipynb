```python
vec = np.array([3,4,5,2])
s = 2

vec*s
```

```python
vec2d = np.array([1,2])
s1 = 2
s2 = .5
s3 = -1

#indexing
vec2d[0]
```

```python
plt.plot([0, vec2d[0]], [0, vec2d[1]], 'bs-', label = 'v')
plt.plot([0, s1*vec2d[0]], [0, s1*vec2d[1]], 'ro-', label = 'v*s1')
plt.plot([0, s2*vec2d[0]], [0, s2*vec2d[1]], 'kp-', label = 'v*s2')
plt.plot([0, s3*vec2d[0]], [0, s3*vec2d[1]], 'g*-', label = 'v*s3')

plt.axis('square')
plt.xlim([-4,4])
plt.ylim([-4,4])
plt.grid()
plt.legend()
plt.show()
```

```python
# assigment
# create two vectors
# plot each vector
# plot v1+v2, v1-v2, v1*4 + v2/2

vec1 = np.array([2,4,5,6])
vec2 = np.array([6,3,2,4])

sp = 2

#indexing
vec1[0]
vec2[0]

plt.plot([0, vec1[0]], [0, vec1[1]], 'bs-', label = 'v1')
plt.plot([0, vec2[0]], [0, vec2[1]], 'ro-', label = 'v2')
plt.plot([0, vec1[0] + vec2[0]], [0, vec1[1] + vec2[1]], 'kp-', label = 'v1+v2')
plt.plot([0, vec1[0] - vec2[0]], [0, vec1[1] - vec2[1]], 'k*-', label = 'v1-v2')
plt.plot([0, 4*vec1[0] + vec2[0]/2], [0, 4*vec1[1] + vec2[1]/2], 'g*-', label = 'v1-v2')

plt.axis('square')
plt.xlim([-4,4])
plt.ylim([-4,4])
plt.grid()
plt.legend()
plt.show()
```
